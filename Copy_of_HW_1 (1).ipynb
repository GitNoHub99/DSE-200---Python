### 4. Write a Python Program to Check if a number has exactly 3 digits(no zeros allowed at the beginning)

* Initialize a variable `num`
* Check and print if the number has 3 digits(no zeros allowed at the beginning)

````python
def has_3_digits(s):
  s = s.strip()  #removes any spaces between quotes
  if s.startswith('-'):   #removes any potential negative signs
    s = s[1:]

  if len(s) != 3: #checks to make sure value is 3 characters
    return False

  if not s.isdigit(): #checks to make sure all characters are digits
    return False

  if s[0] == '0': #checks to make sure no leading zeroes
    return False

  return True

num = "909" # Take input from user

if has_3_digits(num): # Check if the number has exactly 3 digits, w/ no leading zeroes
  print(f"{num} has exactly 3 digits and no leading zeroes")
else:
  print(f"{num} does not have 3 digits and/or has leading zeroes")
````

* Use a for loop to remove consonants(any letter other than a,e,i,o,u)

````python
def removeConsonants(s):
  vowels = "aeiouAEIOU"  # defines all vowels - upper/lower
  result = ""            # creates empty string
  for char in s:         # loops through each char in string
    if char in vowels or not char.isalpha():  #checks each char in string for vowels & not letters.
      result += char     # takes each true condition (not a vowel, not a letter), and adds them to list
  return result
my_str = "Hello, how are you?. I hope you are doing well."

print(removeConsonants(my_str))
````

### 10. Define a class named Triangle that defines the methods `area` and `perimeter`

* The class can be constructed by the length of the sides. Hint - As arguments of `__init__`
* $area = \sqrt{(s(s-a)(s-b)(s-c)}$ where $s=\frac{(a+b+c)}{2}$ is the semi-perimeter
* $perimeter = a +  b + c$
* **Optional** - If there are repeated calls to the `area` and the `perimeter` functions, how would you optimize the code?

````python
class TriangleOptimized:
    def __init__(self, a, b, c):
      self.a = a
      self.b = b
      self.c = c

      self._area = None   #cache variable
      self._perimeter = None  #cache variable

    def area(self):
      if self._area is not None:  #caching logic
        return self._area         #return cached value
      s = (self.a + self.b + self.c) / 2
      self._area = math.sqrt(s * (s - self.a) * (s - self.b) * (s - self.c))
      return self._area

    def perimeter(self):
      if self._perimeter is not None:
        return self._perimeter
      self._perimeter = self.a + self.b + self.c
      return self._perimeter

triangle1 = TriangleOptimized(3, 4, 5)
triangle2 = TriangleOptimized(5, 12, 13)
triangle3 = TriangleOptimized(7, 7, 7)

# Test Triangle 1
print(f"Triangle with sides: {triangle1.a}, {triangle1.b}, {triangle1.c}")
print("Calling area() first time (calculates):", triangle1.area())
print("Calling area() second time (cached):", triangle1.area())
print("Calling perimeter() first time (calculates):", triangle1.perimeter())
print("Calling perimeter() second time (cached):", triangle1.perimeter())

# Test Triangle 2
print(f"\nTriangle with sides: {triangle2.a}, {triangle2.b}, {triangle2.c}")
print("Calling area() first time (calculates):", triangle2.area())
print("Calling area() second time (cached):", triangle2.area())
print("Calling perimeter() first time (calculates):", triangle2.perimeter())
print("Calling perimeter() second time (cached):", triangle2.perimeter())

# Test Triangle 3
print(f"\nTriangle with sides: {triangle3.a}, {triangle3.b}, {triangle3.c}")
print(f"Calling area() first time (calculates): {triangle3.area():.1f}")
print(f"Calling area() second time (cached): {triangle3.area():.1f}")
print("Calling perimeter() first time (calculates):", triangle3.perimeter())
print("Calling perimeter() second time (cached):", triangle3.perimeter())
````

````python
import math
class Triangle:    #creates basic triangle class
  def __init__(self, a, b, c):  #Initialize a triangle with three sides. a, b, c are the lengths of the sides.
    self.a = a
    self.b = b
    self.c = c

  def area(self):   #calculates area
    s = (self.a + self.b + self.c) / 2
    return math.sqrt(s * (s - self.a) * (s - self.b) * (s - self.c))

  def perimeter(self): #calculates perimeter
    return self.a + self.b + self.c

a = 3
b = 4
c = 5
triangle = Triangle(a, b, c)
triangle2 = Triangle(9, 5, 10)

print(f"Triangle with sides: {triangle.a}, {triangle.b}, {triangle.c}")
print("Area of the triangle:", triangle.area())
print("Perimeter of the triangle:", triangle.perimeter())

print(f"Triangle with sides: {triangle2.a}, {triangle2.b}, {triangle2.c}")
print(f"Area of the triangle: {triangle2.area():.1f}")
print(f"Perimeter of the triangle: {triangle2.perimeter()}")
````
